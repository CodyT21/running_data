# Running Data Analytics

This notebook will serve as the starting point for analyzing my running data dating back to 2022. Running data will come from Fitbit and Strava.

```python
# import packages

import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)

# data visualizations
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
strava_activities = pd.read_csv('activities_strava.csv', header=0)
strava_activities.head()

strava_runs = strava_activities.loc[strava_activities['Activity Type'] == 'Run']
print(strava_activities.shape)
print(strava_runs.shape)
```

There are 176 runs tracked in Strava.

```python
strava_runs.info()
```

First all columns that only contain null values will be dropped.

```python
strava_runs = strava_runs.dropna(how='all', axis=1)
strava_runs.info()
```

```python
strava_runs.head()
```

```python
strava_runs.describe(include='all')
```

```python
cols_to_drop = ['Activity ID', 'Activity Type', 'Activity Description', 'Commute', 'Activity Private Note', 'Activity Gear', \
                'Filename', 'Commute.1', 'From Upload', 'Gear', 'Flagged', 'Max Heart Rate.1', 'Prefer Perceived Exertion', \
                'Precipitation Type', 'Weather Ozone', 'Average Grade Adjusted Pace']
trimmed_strava_runs = strava_runs.drop(columns=cols_to_drop, axis=1)
trimmed_strava_runs.head()
```

```python
trimmed_strava_runs.isna().sum()
```

```python
trimmed_strava_runs.loc[trimmed_strava_runs.isna().any(axis=1)]
```

```python
# removed runs that weren't tracked correctly
trimmed_strava_runs = trimmed_strava_runs.dropna(subset=['Max Heart Rate', 'Average Heart Rate'], axis=0)

# fill missing weather data will data from previous day
trimmed_strava_runs = trimmed_strava_runs.fillna(method='bfill', axis=0)
trimmed_strava_runs.isna().sum()
```

```python
# convert column names to lower case and snake case
column_names = {column: column.strip().lower().replace(' ', '_').replace('.1', '') for column in trimmed_strava_runs.columns}
trimmed_strava_runs = trimmed_strava_runs.rename(columns=column_names)
trimmed_strava_runs.head()
```

```python
# convert date and time
trimmed_strava_runs['activity_date'] = pd.to_datetime(trimmed_strava_runs['activity_date'])
trimmed_strava_runs.head()
```

```python
# convert activity_name to a categorical for time of day for the run
trimmed_strava_runs['time_of_day'] = trimmed_strava_runs['activity_name'].str.lower().str.split(' ').str[0]
trimmed_strava_runs['time_of_day'] = trimmed_strava_runs['time_of_day'].astype('category')
trimmed_strava_runs['time_of_day']
```

```python
max_heart_rate = trimmed_strava_runs['max_heart_rate'].quantile(0.99) # potential outlier in the max heart rate column

def heart_rate_zones(max_heart_rate, heart_rate):
    if heart_rate < 0.6 * max_heart_rate:
        return 1
    elif heart_rate < 0.7 * max_heart_rate:
        return 2
    elif heart_rate < 0.8 * max_heart_rate:
        return 3
    elif heart_rate < 0.9 * max_heart_rate:
        return 4
    else:
        return 5

trimmed_strava_runs['average_running_zone'] = trimmed_strava_runs.apply(lambda x: heart_rate_zones(max_heart_rate, x['average_heart_rate']), axis=1).astype('category')
trimmed_strava_runs['average_running_zone']
```
